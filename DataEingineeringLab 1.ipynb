```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("C:/Users/Dell/Downloads/customer_feedback (1).csv")
df1
```

```python
df2=pd.read_csv("C:/Users/Dell/OneDrive/Desktop/machine learning/data engineering lab 1/sale_price.csv")
df2
```

```python
df2['sale_price'] = df2['sale_price'].astype(str).str.replace('$', '', regex=False).astype(float)
df2
```

```python
total_sales=df2.groupby('product_id')['sale_price'].sum().reset_index()
total_sales
```

```python
total_r=df1.groupby('product_id')['sentiment_score'].sum().reset_index()
total_r
```

```python
plt.bar(total_sales['product_id'], total_sales['sale_price'], color='skyblue')
plt.xlabel('Product ID')
plt.ylabel('Total Sale Price')
plt.title('Total Price per Product')
plt.show()
```

```python
plt.bar(total_r['product_id'], total_r['sentiment_score'], color='skyblue')
plt.xlabel('Product ID')
plt.ylabel('Total Product Ratings')
plt.title('Product Sentiment Analysis')
plt.show()
```

```python
sales_product=df2.groupby('product_id')['sale_id'].count().reset_index()
sales_product
```

```python
plt.bar(sales_product['product_id'], sales_product['sale_id'], color='skyblue')
plt.xlabel('Product ID')
plt.ylabel('Total Number Of Sales')
plt.title('Product Sale Analysis')
plt.show()
```

```python
sales_product.rename(columns={'sale_id':'number_of_sales'},inplace=True)
sales_product
```

```python
average_sale=df2.groupby('product_id')['sale_price'].mean().reset_index()
average_sale
```

```python
plt.bar(average_sale['product_id'], average_sale['sale_price'], color='skyblue')
plt.xlabel('Product ID')
plt.ylabel('Sale-price')
plt.title('Average Product Sale Analysis')
plt.show()
```
